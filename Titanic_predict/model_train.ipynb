```python
import pandas as pd
import numpy as np
```

```python
train_feature = pd.read_csv("/Users/wuyijiong/Desktop/kaggle-Titanic-predict/Titanic_predict/data/train_feature.csv")
test_feature = pd.read_csv("/Users/wuyijiong/Desktop/kaggle-Titanic-predict/Titanic_predict/data/test_feature.csv")
```

```python
from sklearn import linear_model
# 用正则取出我们要的属性值
train_df = train_feature.filter(regex='Survived|Age_.*|SibSp|Parch|Fare_.*|Cabin_.*|Embarked_.*|Sex_.*|Pclass_.*')
test_df = test_feature.filter(regex='Survived|Age_.*|SibSp|Parch|Fare_.*|Cabin_.*|Embarked_.*|Sex_.*|Pclass_.*')
train_np = train_df.values
#test_np = test_df.values
```

```python
# y即Survival结果
y_train = train_np[:, 0]
# X即特征属性值
X_train = train_np[:, 1:]
```

```python
# fit到RandomForestRegressor之中
clf = linear_model.LogisticRegression(C=1.0, penalty='l1', tol=1e-6,solver='liblinear')
clf.fit(X_train, y_train)
```

```python
test_predictions = clf.predict(test_df)
test_results =  pd.DataFrame({'PassengerId':test_feature['PassengerId'].values, 'Survived':test_predictions.astype(np.int32)})
```

```python
test_results.to_csv("/Users/wuyijiong/Desktop/kaggle-Titanic-predict/Titanic_predict/data/test_results.csv")
```
