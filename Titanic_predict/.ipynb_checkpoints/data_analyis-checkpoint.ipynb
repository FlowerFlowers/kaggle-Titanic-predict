```python
import numpy as np
import  pandas as pd
import matplotlib.pyplot as plt
```

```python
data_train = pd.read_csv("E:/code/python/Titanic_predict/data/train.csv")
data_train
```

```python
data_train.info()
#age缺失大概20%，cabin缺失80%，embarked两个缺失
```

```python
data_train.describe()
#大概38%人最终获救，平均年龄29岁等......
```

```python
plt.rcParams['font.sans-serif']=['SimHei']#用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False#用来正常显示负号
fig = plt.figure()
fig.set(alpha=0.5)  # 设定图表颜色alpha参数

plt.subplot2grid((2,3),(0,0))             # 在一张大图里分列几个小图
data_train['Survived'].value_counts().plot(kind='bar')# 柱状图 bar针对于categorical变量，hisr针对于numerical变量
plt.title(u"获救情况 (1为获救)") # 标题
plt.ylabel(u"人数")  

plt.subplot2grid((2,3),(0,1))
data_train.Pclass.value_counts().plot(kind="bar")
plt.ylabel(u"人数")
plt.title(u"乘客等级分布")

plt.subplot2grid((2,3),(0,2))
plt.scatter(data_train.Survived, data_train.Age)
plt.ylabel(u"年龄")                         # 设定纵坐标名称
plt.grid(b=True, which='both', axis='y') #设置网格线
plt.title(u"按年龄看获救分布 (1为获救)")


plt.subplot2grid((2,3),(1,0), colspan=2)
data_train.Age[data_train.Pclass == 1].plot(kind='kde')   #kde核密度,就是概率密度曲线
data_train.Age[data_train.Pclass == 2].plot(kind='kde')
data_train.Age[data_train.Pclass == 3].plot(kind='kde')
plt.xlabel(u"年龄")# plots an axis lable
plt.ylabel(u"密度") 
plt.title(u"各等级的乘客年龄分布")
plt.legend((u'头等舱', u'2等舱',u'3等舱'),loc='best') # 设置图例


plt.subplot2grid((2,3),(1,2))
data_train.Embarked.value_counts().plot(kind='bar')
plt.title(u"各登船口岸上船人数")
plt.ylabel(u"人数")  
plt.show()
```

```python
#看看各乘客等级的获救情况
fig = plt.figure()
fig.set(alpha=0.2)  # 设定图表颜色alpha参数

Survived_0 = data_train['Pclass'][data_train.Survived == 0].value_counts()
Survived_1 = data_train.Pclass[data_train.Survived == 1].value_counts()
df=pd.DataFrame({u'获救':Survived_1, u'未获救':Survived_0})
print(df)
df.plot(kind='bar', stacked=True)#stacked:是否堆放在一起
plt.title(u"各乘客等级的获救情况")
plt.xlabel(u"乘客等级") 
plt.ylabel(u"人数") 
plt.show()
```

```python
#看看各性别的获救情况
fig = plt.figure()
fig.set(alpha=0.2)  # 设定图表颜色alpha参数

Survived_m = data_train['Survived'][data_train.Sex == 'male'].value_counts()
Survived_f = data_train['Survived'][data_train.Sex == 'female'].value_counts()
df=pd.DataFrame({u'男性':Survived_m, u'女性':Survived_f})
df.plot(kind='bar', stacked=True)#stacked:堆放
plt.title(u"不同性别的获救情况")
plt.xlabel(u"性别") 
plt.ylabel(u"人数") 
plt.show()
```

```python
#然后我们再来看看各种舱级别情况下各性别的获救情况
fig=plt.figure()
fig.set(alpha=0.65) # 设置图像透明度，无所谓
plt.title(u"根据舱等级和性别的获救情况")

ax1=fig.add_subplot(141)
#女性高级仓获救情况统计
#value_counts默认按照sort=True排序，即值大的排前面
data_train.Survived[data_train.Sex == 'female'][data_train.Pclass != 3].value_counts(sort = True).plot(kind='bar', label="female highclass", color='#FA2479')
ax1.set_xticklabels([u"获救", u"未获救"], rotation=0)
ax1.legend([u"女性/高级舱"], loc='best')
#女性低级仓情况统计
ax2=fig.add_subplot(142, sharey=ax1)#sharey 共享y轴
#sort_index给dataframe按照index递增排序
a=data_train.Survived[data_train.Sex == 'female'][data_train.Pclass == 3].value_counts().sort_index().plot(kind='bar', label='female, low class', color='pink')
ax2.set_xticklabels([u"未获救", u"获救"], rotation=0)
plt.legend([u"女性/低级舱"], loc='best')

ax3=fig.add_subplot(143, sharey=ax1)
data_train.Survived[data_train.Sex == 'male'][data_train.Pclass != 3].value_counts().plot(kind='bar', label='male, high class',color='lightblue')
ax3.set_xticklabels([u"未获救", u"获救"], rotation=0)
plt.legend([u"男性/高级舱"], loc='best')

ax4=fig.add_subplot(144,sharey=ax1)
#value_counts默认降序排列
data_train['Survived'][data_train.Sex =='male'][data_train.Pclass == 3].value_counts().plot(kind='bar',label='male low class')
ax4.legend([u"男性低级仓"],loc='best')
ax4.set_xticklabels([u"未获救",u"获救"])
plt.show()
```

```python
#各登录港口
fig = plt.figure()
survied = data_train['Embarked'][data_train.Survived == 1].value_counts()
unsurvied = data_train['Embarked'][data_train.Survived == 0].value_counts()
df = pd.DataFrame({u"获救人数":survied,u"未获救人数":unsurvied})
print(df)
df.plot(kind="bar")
plt.xlabel(u"港口")
plt.title(u"港口获救人数统计图")
plt.ylabel(u"人数")
plt.show()
```

```python
#堂兄弟/妹，孩子/父母有几人，对是否获救的影响
temp = data_train.groupby(["SibSp","Survived"])
df=pd.DataFrame(temp.count()["PassengerId"])
temp2 = data_train.groupby(["Parch","Survived"])
df2=pd.DataFrame(temp2.count()["PassengerId"])
```

```python
#探索船舱号对结果是否影响
#data_train['Cabin'].value_counts()
fig = plt.figure()
Survived_cabin = data_train.Survived[pd.notnull(data_train.Cabin)].value_counts()
Survived_nocabin = data_train.Survived[pd.isnull(data_train.Cabin)].value_counts()
#tanspose行列转置
df=pd.DataFrame({u"有cabin号":Survived_cabin,u"没有cabin号":Survived_nocabin}).transpose()
df.plot(kind="bar",stacked=True)
plt.title(u"是否有cabin号对获救的影响")
plt.xlabel(u"是否有船舱号")
plt.ylabel(u"人数")
plt.show()
```
